```python
# Document Intelligence transfer PDF to text
from azure.core.credentials import AzureKeyCredential
from azure.ai.formrecognizer import DocumentAnalysisClient

document_analysis_client = DocumentAnalysisClient(endpoint=os.environ["DOCU_INTELLIGENCE_ENDPOINT"], 
                                                  credential=AzureKeyCredential(os.environ["DOCU_INTELLIGENCE_KEY"]))

with open('C:/Users/joann/Downloads/Papers/Math/2311.09988.pdf', "rb") as f:
    poller = document_analysis_client.begin_analyze_document(
        "prebuilt-read", document=f
    )
    not_completed=False
result = poller.result()
```

```python
result.pages
```

```python
page_content = []
for page in result.pages:

    all_lines_content = []
    for line_idx, line in enumerate(page.lines):
        all_lines_content.append(' '.join([word.content for word in line.get_words()]))
    page_content.append({'filename':'2311.09988.pdf',
                         'page_number':page.page_number, 
                         'page_content':' '.join(all_lines_content)})
```

```python
page_content
```

```python
# Embedding
embeddings= []
for content in page_content:
    content['embedding']=openai.Embedding()\
            .create(input=[content['page_content']], engine='text-embedding-ada-002')["data"][0]["embedding"]
```
